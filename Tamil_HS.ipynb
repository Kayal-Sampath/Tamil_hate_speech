```python
train_df = pd.read_csv("train.csv",index_col=False)
train_df.head()
```

```python
dev_df = pd.read_csv("eval.csv",index_col=False)
dev_df.head()
```

```python
test_df = pd.read_csv("test_labels.tsv",index_col=False, delimiter = "\t")
test_df.head()
```

```python
df = train_df.append(dev_df,ignore_index=True)
```

```python
import re

def preprocess(sentence):
    sentence=str(sentence)
    rem_hashtag=re.sub(r'#\S+', '',sentence)
    rem_tag=re.sub(r'@\S+', '',rem_hashtag)
    rem_punc=re.sub(r'[,|:-<>!@#$%()_-]', '',rem_tag)
    rem_url=re.sub(r'http\S+', '',rem_punc)
    return "".join(rem_url)

df['cleanText']=df['text'].map(lambda s:preprocess(s))
df.head(5)
```

```python
!pip install -U sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
text = df['cleanText'].values.tolist()
```

```python
lab_model = SentenceTransformer('sentence-transformers/LaBSE')
lab_emb = lab_model.encode(text, batch_size=50, show_progress_bar=True)
lab_emb.shape
```

```python
X_train = lab_emb
y_train = df['labels'].values.tolist()
```

```python
from sklearn import preprocessing
from sklearn.metrics import classification_report,confusion_matrix


le = preprocessing.LabelEncoder()
le.fit(test_df['category'].tolist())
test_df ['label_le'] = le.transform(test_df['category'].tolist())
y_test = test_df['label_le'].values.tolist()


test_df['cleanText']=test_df['text'].map(lambda s:preprocess(s))
test_text =  test_df['cleanText'].values.tolist()
X_test = lab_model.encode(test_text, batch_size=50, show_progress_bar=True)
```

##Without smote classifiation

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 100)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn import linear_model

clf = linear_model.LogisticRegression()

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(random_state=1, max_iter=300)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

##Using Smote

```python
from imblearn.over_sampling import SMOTE

def smoting(X_tr, y_tr):
  smote = SMOTE(random_state=42)
  X, y = smote.fit_resample(X_tr, y_tr)
  #y = smote.fit_sample()
  return X, y

X_smote, y_smote = smoting(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 100)
clf.fit(X_smote, y_smote)

y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(random_state=1, max_iter=300)

clf.fit(X_smote, y_smote)

y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)

clf.fit(X_smote, y_smote)

y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn import linear_model

clf = linear_model.LogisticRegression()

clf.fit(X_smote, y_smote)

y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

clf = make_pipeline(StandardScaler(), SVC(kernel = "linear", gamma='auto'))

clf.fit(X_smote, y_smote)

y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```
